```python
# Auto run-and-record
# user defined parameter: RUNS - default = 200
# what does this function do: first del all contents under "logs" dir
# -> exec launch.py <RUNS> times
# -> extract time cost and write only one line to log: @ Driver: ElapsedSec: <timecost.4f>
# -> collect to times_release list

import re
import subprocess
import time
from pathlib import Path

RUNS = 300
ROOT = Path('.').resolve()
LOG_DIR = ROOT / 'local_buffer' / 'logs'
LOG_DIR.mkdir(parents=True, exist_ok=True)

# clean logs dir
for p in LOG_DIR.iterdir():
    if p.is_file():
        p.unlink()

_times = []
for i in range(1, RUNS + 1):
    log_path = LOG_DIR / f'launch_auto_{i:03d}_{time.strftime("%Y%m%d_%H%M%S")}.log'
    proc = subprocess.run(['python', 'lifecycle/launch.py'], cwd=ROOT, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True, check=False)
    elapsed = None
    for line in proc.stdout.splitlines():
        m = re.search(r'@ Driver: ElapsedSec: ([0-9.]+)', line)
        if m:
            elapsed = float(m.group(1))
            break
    with log_path.open('w') as f:
        if elapsed is not None:
            f.write(f"@ Driver: ElapsedSec: {elapsed:.4f}\n")
        else:
            f.write("@ Driver: ElapsedSec: NA\n")
    _times.append(elapsed)
    time.sleep(1)

OUTPUT = _times.copy()
print('runs', RUNS)
print('valid', sum(t is not None for t in OUTPUT))
print('min', min(t for t in OUTPUT if t is not None))
print('max', max(t for t in OUTPUT if t is not None))
```

```python
# 3-comp GMM with manual cutoffs (6.25 / 7.25)
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from matplotlib.ticker import MultipleLocator, FuncFormatter

# =========================
# Data
# =========================
arr = np.array([t for t in times_release if t is not None], dtype=float)

# manual cutoffs
c1, c2 = 6.25, 7.25
mask1 = arr < c1
mask2 = (arr >= c1) & (arr < c2)
mask3 = arr >= c2
groups = [arr[mask1], arr[mask2], arr[mask3]]

# =========================
# GMM (3 components)
# =========================
x = arr.reshape(-1, 1)
gmm = GaussianMixture(n_components=3, random_state=42)
gmm.fit(x)

means = gmm.means_.flatten()
vars_ = gmm.covariances_.flatten()
weights = gmm.weights_.flatten()

# =========================
# Solarized Light theme
# =========================
solarized_bg = "#fdf6e3"      # base3
solarized_grid = "#eee8d5"    # base2
solarized_text = "#586e75"    # base01
solarized_hist = "#93a1a1"    # base1

# =========================
# Plot
# =========================
fig, ax = plt.subplots(figsize=(10, 6), dpi=300)
ax.xaxis.set_major_locator(MultipleLocator(0.2))
ax.xaxis.set_major_formatter(FuncFormatter(lambda x, pos: f"{x:.1f}s"))

# background
fig.patch.set_facecolor(solarized_bg)
ax.set_facecolor(solarized_bg)

# =========================
# Top layer: histogram (KEEP AS-IS)
# =========================
ax.hist(
    arr,
    bins=150,
    density=True,
    alpha=0.45,
    color=solarized_hist
)

# =========================
# Boxplots below x-axis:
#   bottom: Overall + infobox
#   above : Group boxplots (no infobox)
# =========================
y_top = ax.get_ylim()[1]  # positive ymax from histogram
box_shift = 0.05 * y_top   # 

# --- Overall stats
overall_mu = arr.mean()
overall_std = arr.std(ddof=1) if arr.size > 1 else 0.0
overall_se = overall_std / np.sqrt(arr.size) if arr.size > 1 else 0.0

# --- Layout (all below x-axis)
overall_extra_drop = 0.05 * y_top   #  Overall =
overall_y = -0.12 * y_top - box_shift - overall_extra_drop
groups_y  = -0.06 * y_top - box_shift

# separator line between group row and overall row
sep_y = (groups_y + overall_y) / 2
xaxis_lw = ax.spines["bottom"].get_linewidth()
ax.axhline(
    y=sep_y,
    color=solarized_text,
    linewidth=xaxis_lw,
    alpha=0.8,
    linestyle='-',
    zorder=0
)

box_h = 0.045 * y_top     # box "height" (in y-data units)

def draw_hbox(data, y, facecolor, edgecolor, flier_marker='o'):
    if data.size == 0:
        return
    ax.boxplot(
        data,
        vert=False,
        positions=[y],
        widths=box_h,
        patch_artist=True,
        showfliers=True,
        flierprops=dict(
            marker=flier_marker,
            markersize=4,
            markerfacecolor=facecolor,
            markeredgecolor=edgecolor,  #   edgecolor
            markeredgewidth=0.6,
            alpha=0.8
        ),
        boxprops=dict(facecolor=facecolor, edgecolor=edgecolor),
        medianprops=dict(color="#b58900", linewidth=1.5),
        whiskerprops=dict(color=edgecolor),
        capprops=dict(color=edgecolor)
    )

# --- 1) Overall boxplot + mean line + infobox (infobox )
draw_hbox(
    arr,
    overall_y,
    facecolor="#eee8d5",
    edgecolor=solarized_text,
    flier_marker='D'   # diamond
)

ax.plot(
    [overall_mu, overall_mu],
    [overall_y - 0.45 * box_h, overall_y + 0.45 * box_h],
    color="#dc322f",
    linewidth=2
)

# bottom boundary line (mirror of sep_y about overall_y)
bottom_sep_y = 2 * overall_y - sep_y
ax.axhline(
    y=bottom_sep_y,
    color=solarized_text,
    linewidth=xaxis_lw,
    alpha=0.8,
    linestyle='-',
    zorder=0
)

# --- Overall mean vertical line (from bottom boundary to x-axis)
ax.vlines(
    overall_mu,
    ymin=bottom_sep_y,
    ymax=0,                 # x-axis (with ticks) at y=0
    color=solarized_text,
    linewidth=0.8,
    linestyle='--',
    alpha=0.7,
    zorder=5
)

# marker at intersection of overall mean line and x-axis
ax.plot(
    overall_mu,
    0,
    marker='o',
    markersize=5,
    markerfacecolor=solarized_bg,   #  = 
    markeredgecolor=solarized_text,    #   = x-axis
    markeredgewidth=1.0,
    zorder=6
)


overall_label = (
    f"Overall\n"
    f"={overall_mu:.4f}\n"
    f"sd={overall_std:.4f}\n"
    f"SE={overall_se:.4f}"
)
#  overall infobox
# ===== Overall infobox (enable) + arrow from mean circle to infobox lower-left =====
# x-axis color (for marker edge & arrow)
xaxis_color = ax.spines["bottom"].get_edgecolor()

# start point: overall mean circle on x-axis
start_data = (overall_mu, 0)

# --- compute 0.2s tick spacing in DISPLAY coords (pixels)
p0 = ax.transData.transform((overall_mu, 0))
p1 = ax.transData.transform((overall_mu + 0.2, 0))
dx_pix = p1[0] - p0[0]

# --- 45 vector with length sqrt(2)*dx_pix -> (dx_pix, dx_pix)
end_disp = (p0[0] + dx_pix, p0[1] + dx_pix)
end_data = ax.transData.inverted().transform(end_disp)

# extra offset to avoid overlap between arrow and infobox (in pixels)
offset_px = (10, 10)   #  (8,8) / (12,12)

end_disp_shifted = (
    end_disp[0] + offset_px[0],
    end_disp[1] + offset_px[1]
)
end_data_shifted = ax.transData.inverted().transform(end_disp_shifted)


# --- place infobox so that its lower-left corner is at arrow end
overall_text = ax.text(
    end_data_shifted[0],
    end_data_shifted[1],
    overall_label,
    ha="left",
    va="bottom",
    fontsize=9,
    color=solarized_text,
    bbox=dict(
        boxstyle="round,pad=0.35",
        fc="white",
        ec=solarized_grid,
        alpha=0.9
    ),
    zorder=8
)


# --- draw once to get accurate bbox, then aim arrow exactly at infobox lower-left
fig.canvas.draw()
renderer = fig.canvas.get_renderer()
bbox = overall_text.get_window_extent(renderer=renderer)
infobox_ll_disp = (bbox.x0, bbox.y0)
infobox_ll_data = ax.transData.inverted().transform(infobox_ll_disp)

# --- arrow from mean circle to infobox lower-left
ax.annotate(
    "",
    xy=infobox_ll_data,      # arrow head at infobox lower-left
    xytext=start_data,       # start at the mean circle (x-axis)
    arrowprops=dict(
        arrowstyle="->",
        color=xaxis_color,   # edge color same as x-axis
        linewidth=1.0,
        shrinkA=0,           # don't shrink from start
        shrinkB=0            # don't shrink from end
    ),
    zorder=7
)


# --- 2) Group boxplots (no infobox)
for g in groups:
    draw_hbox(g, groups_y, facecolor="white", edgecolor=solarized_text)

# Ensure bottom space is visible
ax.set_ylim(bottom_sep_y - 0.6 * box_h, y_top)

# =========================
# GMM component PDFs (KEEP AS-IS)
# =========================
xs = np.linspace(arr.min() - 0.1, arr.max() + 0.1, 1000)

for i in range(3):
    mu_gmm = means[i]
    sigma = np.sqrt(vars_[i])
    w = weights[i]
    y = w * (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(
        -0.5 * ((xs - mu_gmm) / sigma) ** 2
    )
    ax.plot(xs, y, lw=1.0)

# =========================
# Group statistics & labels (TOP LAYER infoboxes: KEEP AS-IS)
# =========================
y_limit = ax.get_ylim()[1]
# unified y position for group infoboxes (visual alignment)
group_infobox_y = y_top * 0.88


for idx, g in enumerate(groups):
    if g.size == 0:
        continue

    mu = g.mean()
    std = g.std(ddof=1) if g.size > 1 else 0.0
    se = std / np.sqrt(g.size) if g.size > 1 else 0.0

    # group mean vertical line (do NOT enter boxplot area)
    ax.vlines(
        mu,
        ymin=sep_y,   # from sep_y upward
        ymax=y_top,
        linestyle='--',
        linewidth=0.75,
        alpha=0.9
    )
    # marker at intersection of group mean line and x-axis
    ax.plot(
        mu,
        0,
        marker='o',
        markersize=4,
        markerfacecolor=solarized_bg,
        markeredgecolor=solarized_text,    #   x-axis
        markeredgewidth=1.0,
        zorder=6
    )


    label_text = (
        f"N={g.size}\n"
        f"={mu:.4f}\n"
        f"sd={std:.4f}\n"
        f"SE={se:.4f}"
    )

    ax.text(
        mu,
        group_infobox_y,
        label_text,
        va='top',
        ha='center',
        fontsize=9,
        color=solarized_text,
        bbox=dict(
            boxstyle='round,pad=0.4',
            fc='white',
            ec=solarized_grid,
            alpha=0.9
        )
    )


# =========================
# Axes styling (Correct)
# =========================
ax.set_title(
    'Gaussian Mixture Modeling of 300 Runtime Measurements with Group-Based Cutoffs',
    color=solarized_text, pad=10
)
ax.set_xlabel('Time cost(sec)', color=solarized_text)

#  ylabel 3  ylab 
ax.set_ylabel("")

# spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
for spine in ['left', 'bottom']:
    ax.spines[spine].set_color(solarized_text)
    
# --- top boundary line (cap the histogram area)
xaxis_lw = ax.spines["bottom"].get_linewidth()
ax.axhline(
    y=y_top,
    color=solarized_text,
    linewidth=xaxis_lw,
    alpha=0.8,
    linestyle='-',
    zorder=0
)


# ticks
ax.tick_params(colors=solarized_text)

# move x-axis to y=0 (between boxplots and hist)
ax.spines["bottom"].set_position(("data", 0))
ax.xaxis.set_ticks_position("bottom")
ax.xaxis.set_label_position("bottom")

# x-label closer
ax.xaxis.set_label_coords(0.5, -0.02)

# left spine only for hist area
ax.spines["left"].set_bounds(0, y_top)

# grid
ax.grid(axis='y', color=solarized_grid, alpha=0.8)
ax.set_yticks([])

# ------------------------------------------------------------
# Custom horizontal y-labels (3 of them), centered & aligned
# ------------------------------------------------------------
trans_y = ax.get_yaxis_transform()  # x in axes coords, y in data coords

x_ylab = -0.02  # leaves a small gap from y-axis; adjust to -0.018/-0.025 if needed
ylab_kw = dict(
    transform=trans_y,
    ha="right",        # right edge stays left of y-axis
    va="center",       # vertically centered at the given y
    rotation=0,        # horizontal text
    color=solarized_text,
    fontsize=10,
    clip_on=False
)

# 1) histogram y-label
ax.text(x_ylab, 0.5 * y_top, "Density &\n histplot", **ylab_kw)

# 2) grouped boxplots y-label
ax.text(x_ylab, groups_y, "Grouped", **ylab_kw)

# 3) overall boxplot y-label
ax.text(x_ylab, overall_y, "Overall", **ylab_kw)

# If left labels get clipped, uncomment:
# plt.subplots_adjust(left=0.14)

plt.tight_layout()
plt.show()
```
